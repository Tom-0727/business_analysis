```python
# 画图
import pandas as pd

name = 'Y'
df = pd.read_csv(f'./ba_buffer/{name}.csv')
df = df[['平台', '功能', '价格']]
df = df.dropna()
print(df.head())
print(df.columns)

df['平台'] = df['平台'].apply(lambda x: x.lower())
z = df['平台'].to_list()
x = df['功能'].to_list()
y = df['价格'].to_list()

print(len(x), len(y), len(z))  
#print(df['Unnamed: 0'].unique())

import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from matplotlib.lines import Line2D

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

# 为每个平台指定一个颜色
platform_colors = {
    'amz': 'blue',
    'wayfair': 'green',
    '独立站': 'red'
}

# 创建一个空列表，用于存储每个点的颜色
colors = [platform_colors[platform] for platform in z]

# 将类别 x 转换为数值位置
unique_x = sorted(set(x))  # 获取唯一的类别并排序
x_positions = np.array([unique_x.index(category) for category in x])  # 将类别映射到位置

# 添加抖动
jitter = np.random.normal(0, 0.1, size=len(x_positions))
x_positions_jittered = x_positions + jitter

# 创建图形并设置大小
plt.figure(figsize=(17, 20))  # 设置图形大小为 10x6 英寸

# 绘制散点图并设置透明度
plt.scatter(x_positions_jittered, y, c=colors, s=100, alpha=0.2)  # 设置透明度为 0.7

# 添加x轴的刻度标签
plt.xticks(range(len(unique_x)), unique_x, rotation=45)
# 添加标签和标题
plt.xlabel('Category', fontsize=12)  # 可以调整字体大小
plt.ylabel('Price', fontsize=12)
plt.title('Scatter Plot of Categories and Prices by Platform', fontsize=14)

# 添加图例
handles = [Line2D([0], [0], marker='o', color='w', label=platform, markersize=10, markerfacecolor=color) 
           for platform, color in platform_colors.items()]
plt.legend(handles=handles, title='Platform', fontsize=10)

# 添加网格
plt.grid(True)
plt.savefig(f'{name}.png', dpi=300, bbox_inches='tight')
# 显示图形
plt.show()
```
